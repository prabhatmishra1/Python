```python
#extract all text and images from the pdf

#extract text

for pageNumber,page in enumerate(pdf.pages(),start=1):
    text = page.getText()
    txt = open(f'report_page{pageNumber}.txt', 'a') 
    txt.writelines(text)
    txt.close()
        
```

```python
#extract images

for pageNumber,page in enumerate(pdf.pages(),start=1):
    for imgNumber, img  in enumerate(page.getImageList(), start=1):
        xref = img[0]
        pix = fitz.Pixmap(pdf, xref)
        
        if pix.n > 4:
            pix = fitz.Pixmap(fitz.csRGB, pix)
        pix.writePNG(f'image_Page{pageNumber}_{imgNumber}.png')
        
```
